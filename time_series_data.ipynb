```python
import pandas as pd

df = pd.read_csv("./cleanData_AA_DL.csv")
df["flightDate"] = pd.to_datetime(df["flightDate"])
df.head()
ts = df.groupby("flightDate")["baseFare"].mean()
```

```python
import matplotlib.pyplot as plt
plt.plot(ts)
```

```python
pd.DataFrame(ts).to_csv("time_series.csv")
```
